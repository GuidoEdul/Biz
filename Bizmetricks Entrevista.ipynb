```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Desktop/datos_arts.csv")
df.head(5)
```

```python
df_agrupado = df.groupby("fecha").agg(
    cantidad_articulos=("articulo", "nunique"),
    total_importe=("importe", "sum")
)
df_agrupado
```

```python
df['SECTOR'] = df['grupo'].apply(lambda x: 'NO COMESTIBLES' if x in ['15 LIMPIEZA DEL HOGAR', '18 PERFUMERIA'] 
                                  else ('PERECEDEROS' if x in ['4 LACTEOS', '9 FIAMBRERIA'] 
                                        else x))
df
```

```python
df[df["SECTOR"] == "PERECEDEROS"]
```

```python
df[df["SECTOR"] == "NO COMESTIBLES"]
```

```python
sector_stats = df.groupby("SECTOR").agg({"precio": "mean", "importe": "sum"})


new_df = pd.DataFrame({
    "Sector": sector_stats.index.repeat(2),
    "Metric": ["Mean Price", "Total Sales"] * len(sector_stats),
    "Value": np.concatenate((sector_stats["precio"].values, sector_stats["importe"].values))
})


wide_df = new_df.pivot(index="Metric", columns="Sector", values="Value")







df.set_index('fecha')
print(df)


wide_df
```
